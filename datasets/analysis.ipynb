```python
import pandas as pd
df1 = pd.read_csv('dim_date.csv')
df2 = pd.read_csv('electric_vehicle_sales_by_makers.csv')
df3 = pd.read_csv('electric_vehicle_sales_by_state.csv')
```

```python
day_by_day = df2.groupby(['date','vehicle_category','maker']).agg({'electric_vehicles_sold':'sum'}).reset_index()
```

```python
day_by_day
```

```python
df3.head()
```

```python
state_by_date = df3.groupby(['date','state','vehicle_category']).agg({'electric_vehicles_sold':'sum','total_vehicles_sold':'sum'}).reset_index()
```

```python
state_by_date
```

```python
df1_2 = pd.merge(left=df1,right=day_by_day,how="inner",on="date")
```

```python
master_df = pd.merge(left=df1_2,right=state_by_date,how="inner",on=["date","vehicle_category"])
```

```python
master_df.rename(columns={'electric_vehicles_sold_x':'electric_vehicles_sold_maker_day','electric_vehicles_sold_y':"total_electric_vehicles_sold_day",'total_vehicles_sold':'total_vehicles_sold_day'},inplace=True)
```
